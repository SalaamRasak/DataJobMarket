```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df_DA_US = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')]
```

```python
df_DA_US.info()
```

```python
sns.set_theme(style='ticks')
sns.barplot(x=df_DA_US['job_location'].value_counts().head(10).values, 
            y=df_DA_US['job_location'].value_counts().head(10).index, 
            orient='h', palette='Blues')
# Add the value annotations
ax = plt.gca()
for i, v in enumerate(df_DA_US['job_location'].value_counts().head(10).values):
    ax.text(v - 40, i, str(v), color='#003366', ha="center", va="center")
plt.xlabel('Number of Job Posted')
plt.ylabel('')
plt.title('Top 10 Job Locations')
plt.show()
```

```python
sns.barplot(x=df_DA_US['company_name'].value_counts().head(10).values, 
            y=df_DA_US['company_name'].value_counts().head(10).index, 
             palette='viridis')
# Add the value annotations
ax = plt.gca()
for i, v in enumerate(df_DA_US['company_name'].value_counts().head(10).values):
    ax.text(v - 40, i, str(v), color='white', ha="center", va="center")
plt.xlabel('Number of Job Posted')
plt.ylabel('')
plt.title('Top 10 Job Posting by Company')
plt.show()
```
